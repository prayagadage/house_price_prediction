### - We have to clean and analyze the Floor, Area Locality, Tenant Preferred
### - Rent and size are left skewed

```python
df[['Floor_Number', 'Total_Floors']] = df['Floor'].str.split(' out of ', expand=True)
```

```python
# Function to clean and convert Floor Number to integer
def clean_floor_number(x):
    if x == 'Ground':
        return 0
    elif 'Upper Basement' in x:
        return -1
    elif 'Lower Basement' in x:
        return -2
    else:
        try:
            return int(x)
        except:
            return None 

# Apply the cleaning function
df['Floor_Number'] = df['Floor_Number'].apply(clean_floor_number)
```

```python
# Convert Total_Floors to numeric (missing values will be NaN)
df['Total_Floors'] = pd.to_numeric(df['Total_Floors'], errors='coerce')
```

```python
df.drop('Floor', axis=1, inplace=True)
```

## Correlation Analysis


```python
# 5. Correlation Matrix
plt.figure(figsize=(10, 8))
numeric_cols = ['BHK', 'Rent', 'Size', 'Bathroom', 'Floor_Number', 'Total_Floors']
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

### 1. Rent Distribution & Skewness


```python
# Rent Skewness and Distribution
print(f"Rent Skewness: {df['Rent'].skew()}")
plt.figure(figsize=(10, 6))
sns.histplot(df['Rent'], kde=True, bins=50)
plt.title('Rent Distribution')
plt.xlabel('Rent')
plt.show()

# Log transformation to visualize better
import numpy as np
plt.figure(figsize=(10, 6))
sns.histplot(np.log1p(df['Rent']), kde=True, bins=50)
plt.title('Log-transformed Rent Distribution')
plt.xlabel('Log(Rent)')
plt.show()
```
